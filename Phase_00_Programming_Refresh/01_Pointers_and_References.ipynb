## 1. Memory Addresses and the Address-of Operator (`&`)

Every variable we declare is stored in a specific location in the computer's memory (RAM). Each of these memory slots has a unique numerical address.
- In C++, we use the **address-of operator (`&`)** to find the memory address where a variable is stored.
- Addresses are typically printed in hexadecimal format (e.g. `0x7ffee3bf87bc`).

```python
%%writefile temp.cpp
#include <iostream>
using namespace std;

int main() {
    int x = 42;
    cout << "Value of x:           " << x << "\n";
    cout << "Memory Address of x:  " << &x << "\n"; // Prints hex address
    return 0;
}
```

## 2. Pointers (`*`) — Storing Addresses and Dereferencing

A **pointer** is a variable that stores the memory address of another variable.
- **Declaration:** Use an asterisk after the type, e.g. `int* ptr;` (declares a pointer to an integer).
- **Initialization:** Assign it a memory address, e.g. `ptr = &x;`.
- **Dereferencing:** Use the asterisk `*` on a pointer to read or write the value stored at that address, e.g. `*ptr = 100;` changes `x` to 100.
- **Null Safety:** If a pointer doesn't point to any address, initialize it to `nullptr` to avoid pointing to random memory (dangling pointers).

```python
%%writefile temp.cpp
#include <iostream>
using namespace std;

int main() {
    int x = 42;
    int* ptr = &x; // ptr stores the address of x
    
    cout << "ptr stores address:       " << ptr << "\n";
    cout << "Value ptr points to (*ptr): " << *ptr << "\n"; // Dereferencing
    
    *ptr = 99; // Modifies x directly
    cout << "New value of x:           " << x << "\n";
    
    int* nullPtr = nullptr;
    if (nullPtr == nullptr) {
        cout << "nullPtr is safe and null.\n";
    }
    return 0;
}
```

```python
!g++ -std=c++17 -O2 -Wall -o temp temp.cpp && ./temp
```

## 3. References (`&`) — Creating Aliases

A **reference** is an alias—another name—for an existing variable.
- **Declaration:** Use an ampersand in the type declaration, e.g. `int& ref = x;`.
- **Rules:**
  1. A reference must be initialized when declared.
  2. It cannot be null.
  3. Once bound to a variable, it cannot be changed to refer to another variable.
- **Benefit:** No dereferencing syntax (no `*`) is required. Modifying the reference modifies the original variable directly.

```python
%%writefile temp.cpp
#include <iostream>
using namespace std;

int main() {
    int x = 42;
    int& ref = x; // ref is now an alias for x
    
    ref = 100; // Modifies x
    cout << "Value of x:   " << x << "\n";
    cout << "Value of ref: " << ref << "\n";
    return 0;
}
```

```python
!g++ -std=c++17 -O2 -Wall -o temp temp.cpp && ./temp
```

## 4. Function Parameter Passing — Value, Pointer, and Reference

How we pass variables to functions determines whether the function works on copies or modifies the caller's variables:
1. **Pass-by-Value:** The function receives a copy. Modifications inside the function do not affect the caller.
2. **Pass-by-Pointer:** Passes the memory address. The function can modify the original variable using dereferencing (`*`). Can accept `nullptr`.
3. **Pass-by-Reference:** Passes an alias. Modifying the parameter modifies the original variable directly. Highly readable and safe (cannot be null).

```python
%%writefile temp.cpp
#include <iostream>
using namespace std;

void setToZeroValue(int val) {
    val = 0; // Only changes local copy
}

void setToZeroPointer(int* ptr) {
    if (ptr != nullptr) {
        *ptr = 0; // Changes original variable via address
    }
}

void setToZeroReference(int& ref) {
    ref = 0; // Changes original variable directly
}

int main() {
    int a = 10, b = 20, c = 30;
    
    setToZeroValue(a);
    cout << "Pass-by-Value a:     " << a << "\n"; // Expected: 10 (unchanged)
    
    setToZeroPointer(&b);
    cout << "Pass-by-Pointer b:   " << b << "\n"; // Expected: 0 (changed)
    
    setToZeroReference(c);
    cout << "Pass-by-Reference c: " << c << "\n"; // Expected: 0 (changed)
    
    return 0;
}
```

```python
!g++ -std=c++17 -O2 -Wall -o temp temp.cpp && ./temp
```

## 5. Checkpoint & Swap Code Example

### Checkpoint Questions
1. What is the difference between `int* ptr` and `*ptr`?
2. Why must a reference be initialized at the time of declaration?
3. When would you prefer Pass-by-Pointer over Pass-by-Reference?

### Answer Key
1. `int* ptr` declares a pointer variable named `ptr`. `*ptr` dereferences the pointer to access the value stored at the address `ptr` holds.
2. A reference is not a separate variable; it is just an alias for an existing variable. Therefore, it must know who it is aliasing immediately.
3. Prefer Pass-by-Pointer if you want to support optional parameters (by passing `nullptr`) or if you need to rebind the pointer to point to something else inside the function.

### Swap Code Example
A classic exercise to demonstrate pointers vs references is swapping two integers.

```python
%%writefile temp.cpp
#include <iostream>
using namespace std;

// Swapping using pointers
void swapPointers(int* x, int* y) {
    int temp = *x;
    *x = *y;
    *y = temp;
}

// Swapping using references
void swapReferences(int& x, int& y) {
    int temp = x;
    x = y;
    y = temp;
}

int main() {
    int a = 5, b = 10;
    cout << "Original values: a = " << a << ", b = " << b << "\n";
    
    swapPointers(&a, &b);
    cout << "After pointer swap: a = " << a << ", b = " << b << "\n"; // a = 10, b = 5
    
    swapReferences(a, b);
    cout << "After reference swap: a = " << a << ", b = " << b << "\n"; // a = 5, b = 10
    
    return 0;
}
```
